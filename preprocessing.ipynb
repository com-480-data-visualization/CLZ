```python
import pandas as pd
```

```python
# Read a CSV file in the same directory
df2026 = pd.read_csv('2026 QS World University Rankings.csv', 
                 sep=';', 
                 header=2, 
                 decimal=',')
df2026.drop(columns=['Index'], inplace=True)

df2026.columns = df2026.columns.str.lower()
# Display the first 5 rows
print(df2026.head())
```

```python
df2025 = pd.read_csv('2025 QS World University Rankings.csv', 
                 sep=';', 
                 header=3, 
                 decimal=',')
df2025.drop(columns=['INDEX'], inplace=True)
df2025.columns = df2025.columns.str.lower()
df2025.rename(columns={'rank display': 'rank', 
                       'rank display2': 'previous rank',
                       'institution': 'name',}, inplace=True)
# Display the first 5 rows
print(df2025.head())
```

```python
df2024 = pd.read_csv('2024 QS World University Rankings.csv', 
                 sep=';', 
                 header=3, 
                 decimal=',')
df2024.columns = df2024.columns.str.lower()
df2024.rename(columns={'rank display': 'rank', 
                       'rank display2': 'previous rank',
                       'institution': 'name',}, inplace=True)
# Display the first 5 rows
print(df2024.head())
```

```python
df2023 = pd.read_csv('2023 QS World University Rankings.csv', 
                 sep=';', 
                 header=3, 
                 decimal=',')
df2023.columns = df2023.columns.str.lower()
df2023.drop(columns=['age band'], inplace=True)
df2023.rename(columns={'rank display': 'rank', 
                       'rank display2': 'previous rank',
                       'institution': 'name',
                       'score scaled': 'overall score'}, inplace=True)
# Display the first 5 rows
print(df2023.head())
```

```python
df2026.to_csv('cleaned_qs_ranking_2026.csv', index=False, encoding='utf-8-sig')
df2025.to_csv('cleaned_qs_ranking_2025.csv', index=False, encoding='utf-8-sig')
df2024.to_csv('cleaned_qs_ranking_2024.csv', index=False, encoding='utf-8-sig')
df2023.to_csv('cleaned_qs_ranking_2023.csv', index=False, encoding='utf-8-sig')
```
